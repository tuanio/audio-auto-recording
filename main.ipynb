```python
import nltk
from nltk.tokenize import sent_tokenize
import random
import os
import IPython
import pandas as pd
import glob
```

```python
df = pd.read_csv('transcripts.txt', sep='\t')
```

```python
corpus = open('deepwork_raw.txt', 'r', encoding='utf-8').read()
corpus = corpus.replace('\n', ' ')
sentences = sent_tokenize(corpus)
```

```python
# chọn câu ngẫu nhiên
sent = random.choice(sentences)
print('='*20)
print(sent)
print('='*20)
```

```python
# lúc nào nên record
%run recorder.py
IPython.display.Audio(filename="foo.mp3", rate=44100)
```

```python
# có lưu hay không
idx = len(glob.glob('tuan/*.mp3')) + 1 
filename = f'tuan_{idx}'
df = df.append(dict(name=filename, transcript=sent), ignore_index=True)
os.rename('foo.mp3', f'tuan/{filename}.mp3')
```
